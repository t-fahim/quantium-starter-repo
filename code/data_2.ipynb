```python
import os
import pandas as pd
```

```python
BASE_DIR = os.getcwd()
# BASE_DIR
file_path = os.path.join(BASE_DIR,"..","data","daily_sales_data_2.csv")
df = pd.read_csv(file_path)
df.head(5)
```

```python
pink_morsel_sales = df.query("product == 'pink morsel'").copy()
pink_morsel_sales
```

```python
pink_morsel_sales.info()
```

```python
pink_morsel_sales.dropna()
```

```python
pink_morsel_sales["price"] = pink_morsel_sales["price"].astype("str")
pink_morsel_sales.info()
```

```python
pink_morsel_sales["price"] = pink_morsel_sales["price"].str.replace('$','',regex=False).astype("float")
```

```python
pink_morsel_sales.info()
```

```python
pink_morsel_sales
```

```python
pink_morsel_sales['sales'] = pink_morsel_sales.eval('price * quantity')
pink_morsel_sales
```

```python
pink_morsel_sales = pink_morsel_sales[['sales','date','region']]
```

```python
pink_morsel_sales
```

```python
pink_morsel_sales.info()
```

```python
file_path = os.path.join(BASE_DIR,"..","data","daily_sales_data_2_modified.csv")
pink_morsel_sales.to_csv(file_path,index=False)
```
